# üß¨ Genetik Algoritma ile Ama√ß Fonksiyonu Optimizasyonu

## üåû Senaryo: G√ºne≈ü Paneli Yerle≈üiminde Optimum Eƒüim ve Y√∂n

Bir belediye, g√ºne≈ü enerjisi sistemi kurarken panellerin eƒüimini ve y√∂n√ºn√º optimize etmek istemektedir. Ama√ß, toplam enerji verimini maksimize ederken fiziksel kurulum ve y√∂nlenme kƒ±sƒ±tlarƒ±nƒ± saƒülamaktƒ±r.

### üéØ Ama√ß Fonksiyonu
Toplam enerji verimini hesaplayan form√ºl:

$$y = 6x_1 + 4x_2 - 0.1x_1^2$$

### üìä Deƒüi≈ükenler

* $x_1$: **Eƒüim a√ßƒ±sƒ± (¬∞)** $\rightarrow [10, 45]$
* $x_2$: **G√ºney y√∂n√ºne g√∂re sapma (¬∞)** $\rightarrow [0, 90]$

### ‚ö†Ô∏è Kƒ±sƒ±tlar

Projenin fiziksel sƒ±nƒ±rlarƒ± ≈üu ≈üekildedir:

1.  **Fiziksel Sƒ±nƒ±r:**
    $$x_1 + 0.5x_2 \leq 60$$

2.  **Minimum Y√∂n ≈ûartƒ±:**
    $$x_2 \geq 15$$

> **Not:** Bu √ßalƒ±≈üma, yukarƒ±da tanƒ±mlanan optimizasyon problemini **Genetik Algoritma** kullanarak √ß√∂zmeyi ama√ßlamaktadƒ±r.

## 1. Gerekli K√ºt√ºphanelerin ƒ∞√ße Aktarƒ±lmasƒ±
Bu b√∂l√ºmde matematiksel i≈ülemler, rastgele sayƒ± √ºretimi ve sonu√ßlarƒ±n g√∂rselle≈ütirilmesi i√ßin gerekli k√ºt√ºphaneler tanƒ±mlanmƒ±≈ütƒ±r.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

## 2. Genetik Algoritma Parametreleri
Genetik algoritmanƒ±n temel √ßalƒ±≈üma parametreleri (pop√ºlasyon, nesil sayƒ±sƒ±, mutasyon oranƒ± vb.) burada belirlenir.

```python
# ================================
# 1Ô∏è‚É£ ALGORƒ∞TMA AYARLARI
# ================================

POPULASYON_BOYUTU = 50      # Aynƒ± anda ka√ß √ß√∂z√ºm denenecek
NESIL_SAYISI = 100          # Ka√ß kere iyile≈ütirme yapƒ±lacak
MUTASYON_IHTIMALI = 0.2     # Bir √ß√∂z√ºm√ºn bozulma ihtimali (%20)
MUTASYON_BUYUKLUGU = 2.0    # Bozulursa en fazla ka√ß derece oynasƒ±n
CEZA_KATSAYISI = 50         # Kƒ±sƒ±t ihlali varsa ceza
```

## 3. Karar Deƒüi≈ükenlerinin Sƒ±nƒ±rlarƒ± ve Ama√ß Fonksiyonu
Algoritmanƒ±n arama uzayƒ±nƒ± belirleyen sƒ±nƒ±rlar ve maksimize edilecek enerji fonksiyonu tanƒ±mlanmƒ±≈ütƒ±r.

```python
# ================================
# 2Ô∏è‚É£ DEƒûƒ∞≈ûKEN SINIRLARI [min, max]
# ================================
LIMITLER = np.array([
    [10, 45],   # eƒüim
    [0, 90]     # y√∂n
])

# ================================
# 3Ô∏è‚É£ AMA√á FONKSƒ∞YONU
# ================================
def amac_fonksiyonu(cozum):
    """
    cozum = [x1, x2]
    x1 = eƒüim
    x2 = y√∂n
    """
    egim = cozum[0]
    yon = cozum[1]

    enerji = 6*egim + 4*yon - 0.1*(egim**2)
    return enerji

# ================================
# 4Ô∏è‚É£ KISIT KONTROL√ú
# ================================
def kisit_kontrol(cozum):
    ihlal = 0
    egim = cozum[0]
    yon = cozum[1]

    # Fiziksel sƒ±nƒ±r (x1 + 0.5x2 <= 60)
    if egim + 0.5*yon > 60:
        ihlal += 1

    # Minimum y√∂n ≈üartƒ± (x2 >= 15)
    if yon < 15:
        ihlal += 1

    return ihlal

# ================================
# 5Ô∏è‚É£ UYGUNLUK (FITNESS) -> Kazan√ß - Ceza
# ================================
def uygunluk_hesapla(cozum):
    enerji = amac_fonksiyonu(cozum)
    ceza = kisit_kontrol(cozum) * CEZA_KATSAYISI
    return enerji - ceza
```

## 4. Genetik Operat√∂rler (Se√ßim, √áaprazlama, Mutasyon)
Evrimsel s√ºrecin i≈ülemesi i√ßin gereken fonksiyonlar:
* **Rulet Se√ßimi:** ƒ∞yi bireylerin se√ßilme ≈üansƒ±nƒ± artƒ±rƒ±r.
* **√áaprazlama:** Ebeveynlerden yeni bireyler √ºretir.
* **Mutasyon:** √áe≈üitliliƒüi artƒ±rmak i√ßin genleri rastgele deƒüi≈ütirir.

```python
# ================================
# 6Ô∏è‚É£ SE√áƒ∞M (RULET)
# ================================
def rulet_secimi(populasyon, uygunluklar):
    min_deger = np.min(uygunluklar)
    # Negatif uygunluk deƒüerlerini pozitife √ßekmek i√ßin kaydƒ±rma
    if min_deger < 0:
        uygunluklar = uygunluklar - min_deger + 1

    olasiliklar = uygunluklar / np.sum(uygunluklar)
    secilen_index = np.random.choice(len(populasyon), p=olasiliklar)
    return populasyon[secilen_index]

# ================================
# 7Ô∏è‚É£ √áAPRAZLAMA
# ================================
def caprazlama(anne, baba):
    # [anne_eƒüim, baba_y√∂n] ve [baba_eƒüim, anne_y√∂n]
    cocuk1 = np.array([anne[0], baba[1]])
    cocuk2 = np.array([baba[0], anne[1]])
    return cocuk1, cocuk2

# ================================
# 8Ô∏è‚É£ MUTASYON
# ================================
def mutasyon(cozum):
    yeni = cozum.copy()
    for i in range(2):
        if np.random.rand() < MUTASYON_IHTIMALI:
            degisim = np.random.uniform(-MUTASYON_BUYUKLUGU, MUTASYON_BUYUKLUGU)
            yeni[i] += degisim
            # Sƒ±nƒ±rlarƒ±n dƒ±≈üƒ±na √ßƒ±karsa kƒ±rp (clip)
            yeni[i] = np.clip(yeni[i], LIMITLER[i,0], LIMITLER[i,1])
    return yeni
```

## 5. Ana D√∂ng√º ve Sonu√ßlar
Algoritmanƒ±n √ßalƒ±≈ütƒ±rƒ±ldƒ±ƒüƒ± ve sonu√ßlarƒ±n grafik ile g√∂sterildiƒüi ana blok.

```python
# ================================
# 9Ô∏è‚É£ ANA ALGORƒ∞TMA
# ================================

def genetik_algoritma():

    print("\n--- GENETƒ∞K ALGORƒ∞TMA BA≈ûLATILDI ---")
    print(f"Pop√ºlasyon: {POPULASYON_BOYUTU}, Nesil Sayƒ±sƒ±: {NESIL_SAYISI}\n")

    populasyon = np.random.uniform(
        [10, 0],  #eƒüim en az 10 - y√∂n en az 0
        [45, 90], #eƒüim en fazla 45 - y√∂n en fazla 90
        (POPULASYON_BOYUTU, 2)
    )

    en_iyi_degerler = []

    for nesil in range(NESIL_SAYISI):

        uygunluklar = np.array([
            uygunluk_hesapla(b) for b in populasyon
        ])

        en_iyi = populasyon[np.argmax(uygunluklar)]  #np.argmax ‚Üí en b√ºy√ºk deƒüerin indeksi YANƒ∞    POP√úLASYONDAKƒ∞ ‚Üí en iyi eƒüim,y√∂n
        en_iyi_degerler.append(np.max(uygunluklar))

        # üîπ NESƒ∞L RAPORU (SADECE EKLEME)
        if (nesil + 1) % 10 == 0:
            print(
                f"Nesil {nesil+1}: "
                f"En ƒ∞yi Puan = {np.max(uygunluklar):.2f} | "
                f"Eƒüim: {en_iyi[0]:.2f}, Y√∂n: {en_iyi[1]:.2f}"
            )

        yeni_pop = [en_iyi]

        while len(yeni_pop) < POPULASYON_BOYUTU:
            anne = rulet_secimi(populasyon, uygunluklar)
            baba = rulet_secimi(populasyon, uygunluklar)

            cocuk1, cocuk2 = caprazlama(anne, baba)

            cocuk1 = mutasyon(cocuk1)
            cocuk2 = mutasyon(cocuk2)

            yeni_pop.extend([cocuk1, cocuk2])

        populasyon = np.array(yeni_pop[:POPULASYON_BOYUTU])

    print("\n--- EN ƒ∞Yƒ∞ √á√ñZ√úM ---")
    print("En iyi √ß√∂z√ºm:", en_iyi)
    print("Enerji:", amac_fonksiyonu(en_iyi))
    print("Fitness:", uygunluk_hesapla(en_iyi))

    plt.plot(en_iyi_degerler)
    plt.title("Genetik Algoritma - Uygunluk (Fitness) Deƒüi≈üimi")
    plt.xlabel("Nesil Sayƒ±sƒ±")
    plt.ylabel("Uygunluk Deƒüeri (Fitness)")
    plt.show()

genetik_algoritma()
```
